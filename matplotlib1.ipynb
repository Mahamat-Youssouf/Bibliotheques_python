## Surface représentant une fonction à deux variables 
####   f:(x,y) ↦ cos(x)+sin(y)
####   L'ensemble des points M(x;y;z) de l'espace tels que z=f(x,y) est une surface représentant la fonction f. z=f(x,y) est une équation de la surface.

## EXEMPLE 1

```python
def f(x,y):
    return np.cos(x) + np.sin(y)
```

```python

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
X, Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.figure(figsize = (12,8))
ax2 = plt.axes(projection = '3d')
ax2.plot_surface(X,Y,Z, cmap = 'plasma')

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')
```

## EXEMPLE 2

```python
def g(x,y):
    r = np.sqrt(x**2 + y**2)
    return np.sin(r)

X = np.arange(-5,5,0.2)
Y = np.arange(-5,5,0.2)
X,Y = np.meshgrid(X,Y)
Z = g(X,Y)

plt.figure(figsize = (12,8))
ax4 = plt.axes(projection = '3d')
ax4.plot_surface(X,Y,Z, cmap = 'viridis')

ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_zlabel('z')
```

## Exemple 3

```python
def h(x,y):
    return 10 - (x**2 + y**2)

X = np.arange(-5,5,0.2)
Y = np.arange(-5,5,0.2)
X,Y = np.meshgrid(X,Y)

Z = h(X,Y)

plt.figure(figsize = (12,8))
ax4 = plt.axes(projection = '3d')
ax4.plot_surface(X,Y,Z, cmap = 'viridis')

ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_zlabel('z')
```

## Tracé filaire: plot_wireframe()

```python
def h(x,y):
    return 10 - (x**2 + y**2)

X = np.arange(-5,5,0.2)
Y = np.arange(-5,5,0.2)
X,Y = np.meshgrid(X,Y)

Z = h(X,Y)


plt.figure(figsize = (12,8))
ax5 = plt.axes(projection = '3d')
ax5.plot_wireframe(X,Y,Z, cmap = 'viridis')

ax5.set_xlabel('x')
ax5.set_ylabel('y')
ax5.set_zlabel('z')
```
