```python
table_frequence = pd.crosstab(student.sex, 'frequence', normalize = True)
```

```python
table_frequence
```

```python
table_pourcentage = table_frequence.assign(frequence = 100*table_frequence.frequence)
```

```python
table_pourcentage
```

```python
table_frequence = pd.crosstab(student.sex, 'frequence', normalize = True)
table_pourcentage = table_frequence.assign(sex = table_frequence.index, frequence = 100*table_frequence.frequence)
```

```python
sns.barplot(y = 'sex', x = 'frequence', data = table_pourcentage)
```
