# Corr√©lations de variables

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
iris['variety'].unique()
```

```python
correlation = iris.corr()
```

```python
correlation
```

```python
sns.heatmap(correlation)
```

```python
sns.heatmap(correlation, annot = True, cmap = 'coolwarm')
```

```python
sns.heatmap(correlation, annot = True, cmap = 'viridis')
```

```python
sns.heatmap(correlation, annot = True, cmap = 'seismic')
```

```python
sns.heatmap(correlation, annot = True, cmap = 'seismic', linewidths = 4, linecolor = 'y')
```

# Repr√©sentation d‚Äôun tableau crois√© dynamique

```python
airPassagers = pd.read_csv('vols.csv')
```

```python
airPassagers.head()
```

```python
airPassagers['year'].unique()
```

```python
ap = airPassagers.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
ap
```

```python
sns.heatmap(ap)
```

```python
sns.heatmap(ap, linewidths = 2, linecolor = 'y', cmap = 'seismic')
```
