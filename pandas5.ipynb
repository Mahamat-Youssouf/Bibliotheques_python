# Traitement des valeurs manquantes dans un DataFrame 

```python
import numpy as np
import pandas as pd
```

```python
data = { 'Nom': ['Christophe', 'Clara','Lina', 'Amine', 'Amandine'],
        'Maths': [13,10,np.nan,11,np.nan],
        'Physique': [13,13,np.nan,11,12],
        'Informatique': [18,12,np.nan,18,np.nan],
        'Anglais': [np.nan,np.nan,np.nan,np.nan,np.nan]
    }
```

```python
notes = pd.DataFrame(data)
```

```python
notes
```

```python
notes.isnull()
```

```python
notes.notnull()
```

```python
notes
```

```python
notes.dropna()
```

```python
notes
```

```python
notes.dropna(axis = 1, how = 'all')
```

```python
notes
```

```python
notes.dropna(thresh = 1)
```

```python
notes
```

```python
notes.fillna(0)
```

```python
notes.fillna(method = 'ffill', axis = 0)
```
