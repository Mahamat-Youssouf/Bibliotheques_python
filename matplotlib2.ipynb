```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = np.random.normal(0, 1, 100)
```

```python
x
```

```python

plt.figure(figsize = (10,8))
ax = plt.axes(projection = '3d')
ax.scatter3D(x,y,z, s = 40)

ax.set_xlabel('axe x')
ax.set_ylabel('axe y')
ax.set_zlabel('axe z')
```

```python
val_x = np.linspace(0, 20, 200)
val_y = np.cos(2*val_x)
val_z = np.sin(2*val_x)

plt.figure(figsize = (10,8))
ax2 = plt.axes(projection = '3d')
ax2.plot3D(val_x,val_y,val_z, lw = 3, color = 'r')
```

```python
def f(x,y):
    return np.cos(x)+np.sin(y)

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

X, Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.figure(figsize = (12,8))
ax3 = plt.axes(projection = '3d')
ax3.plot_surface(X,Y,Z, cmap = 'plasma')

ax3.set_xlabel('axe x')
ax3.set_ylabel('axe y')
ax3.set_zlabel('axe z')
```

```python
%matplotlib
def h(x,y):
    return 10 - (x**2 + y**2)


plt.figure(figsize = (12,8))
X = np.arange(-5,5, 0.2)
Y = np.arange(-5,5, 0.2)
X,Y = np.meshgrid(X,Y)
Z = h(X,Y)

ax4 = plt.axes(projection = '3d')
ax4.plot_surface(X,Y,Z, cmap = 'viridis')

ax4.set_xlabel('axe x')
ax4.set_ylabel('axe y')
ax4.set_zlabel('axe z')
```

```python

def h(x,y):
    return 10 - (x**2 + y**2)


plt.figure(figsize = (12,8))
X = np.arange(-5,5, 0.2)
Y = np.arange(-5,5, 0.2)
X,Y = np.meshgrid(X,Y)
Z = h(X,Y)

ax4 = plt.axes(projection = '3d')
ax4.plot_wireframe(X,Y,Z, cmap = 'viridis')

ax4.set_xlabel('axe x')
ax4.set_ylabel('axe y')
ax4.set_zlabel('axe z')
```
